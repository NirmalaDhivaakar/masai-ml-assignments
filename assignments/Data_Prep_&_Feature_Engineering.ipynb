Import Libraries

```python
import pandas as pd
from sklearn.preprocessing import LabelEncoder, RobustScaler
```

Create the Dataset

```python
data = {
    'Gender': ['Male', 'Female', 'Female', 'Male', 'Female'],
    'City': ['Mumbai', 'Delhi', 'Chennai', 'Mumbai', 'Delhi'],
    'Size': ['Small', 'Large', 'Medium', 'Large', 'Small'],
    'Age': [25, 45, 32, 60, 28],
    'Fare': [15, 300, 85, 450, 20]
}

df = pd.DataFrame(data)

df
```

Task 1 — Categorical Encoding

Gender → Label Encoding

```python
le = LabelEncoder()
df['Gender'] = le.fit_transform(df['Gender'])

df
```

City → One-Hot Encoding (drop one column)

```python
if 'City' in df.columns:
    df = pd.get_dummies(df, columns=['City'], drop_first=True)
```

Size → Ordered Label Encoding

Correct order:
Small = 0
Medium = 1
Large = 2

```python
size_mapping = {
    'Small': 0,
    'Medium': 1,
    'Large': 2
}

df['Size'] = df['Size'].map(size_mapping)
```

# Convert City Columns to 0/1

```python
df[['City_Delhi','City_Mumbai']] = df[['City_Delhi','City_Mumbai']].astype(int)
```

Task 2 — Feature Scaling

Because Fare contains an outlier (450), we use RobustScaler.

RobustScaler uses median and IQR, making it less sensitive to outliers.

```python
scaler = RobustScaler()

df[['Age','Fare']] = scaler.fit_transform(df[['Age','Fare']])
```

Final DataFrame

```python
df
```

# RobustScaler is used because the Fare column contains an outlier (450).
# RobustScaler scales features using the median and interquartile range (IQR),
# making it less sensitive to extreme values compared to StandardScaler or MinMaxScaler.
